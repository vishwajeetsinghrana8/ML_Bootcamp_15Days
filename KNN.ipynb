```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/MY LENOVO/Desktop/Social_Network_Ads.csv')
data.head()
```

```python
len(data)
```

```python
data_x = data.iloc[:,[2,3]].values
data_y = data.iloc[:,4].values
```

```python
data_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x,data_y, test_size=0.25, random_state=0)
```

```python
len(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s_c = StandardScaler()
X_train = s_c.fit_transform(X_train)
X_test = s_c.transform(X_test)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
cls = KNeighborsClassifier(n_neighbors=19, metric='minkowski', p=2)
cls.fit(X_train,y_train)
```

```python
y_pred = cls.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
c_m = confusion_matrix(y_test,y_pred)
c_m
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred)*100)
```
